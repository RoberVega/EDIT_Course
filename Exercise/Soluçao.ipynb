```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("List of all wells_2022.25.01.xlsx", header=1, index_col="Well Name")

display(data.head(5))

display(data.tail(10))
```

```python
data.rename(columns = {'Rig Release \n(Data de Abandono do Rig)':'Data de Abandono do Rig',
                      'Spud Date \n(Data de perfuração)':'Data de perfuração',
                      'Drillers TD \n(Profundidade da perfuração)':'Profundidade da perfuração'},
            inplace = True)
```

```python
data.loc['GENGIBRE-P02',"Longitude"] = np.nan
data.loc['GENGIBRE-P02']
```

```python
data.nunique(dropna=False)
```

```python
data = data.drop(columns='Country')
data.nunique()
```

```python
print_columns=["Units", "Well Type", "Initial Class", "Terrains", "Basin", "Block", "Field"]
for col in print_columns:
    print("\n")
    print(data[col].value_counts(dropna=False))
```

```python
unit_dict = {"METERS":1, "FEET": 0.3048}

data['Units'].map(unit_dict)
data["Profundidade em metros"] = data['Units'].map(unit_dict)*data["Profundidade da perfuração"]
data[data["Profundidade em metros"]<10]
data[data["Profundidade da perfuração"]<10]
```

```python
bol_series=data["UWI (Unique Well Identifier)"].duplicated()
display(bol_series)
data["UWI (Unique Well Identifier)"][bol_series]
```

```python
%matplotlib notebook

data["Profundidade em metros"].plot.hist(xlim=(0,7500),bins=50)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [5, 5]

data.groupby("Initial Class")["Profundidade em metros"].mean().reset_index().plot.bar(x="Initial Class", y="Profundidade em metros");
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [10, 10]

data.groupby("Operator")["Profundidade em metros"].mean().reset_index().plot.barh(x="Operator", y="Profundidade em metros");
```
